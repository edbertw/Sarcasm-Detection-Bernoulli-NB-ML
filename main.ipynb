```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB

dat = pd.read_json("https://raw.githubusercontent.com/amankharwal/Website-data/master/Sarcasm.json", lines=True)
dat.head()
```

```python
dat["is_sarcastic"] = dat["is_sarcastic"].map({0: "No Sarcasm", 1: "Sarcasm"})
dat.head()
```

```python
dat = dat[["headline", "is_sarcastic"]]
X = np.array(dat["headline"])
y = np.array(dat["is_sarcastic"])

Vectorizer = CountVectorizer()
X = Vectorizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
model = BernoulliNB(alpha = 0.5, fit_prior = False)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
input_text = input("Enter a Text: ")
test_run = Vectorizer.transform([input_text]).toarray()
out = model.predict(test_run)
print(out[0])
```
